```
from PIL import Image;
import os;
import cv2;
from google.colab.patches import cv2_imshow

def inp_1(n):
  f=[];
  for i in range(n):
    j=i+1;
    print("Enter image %d path with extension" %(j));
    p=input('');
    img=cv2.imread(p);
    s=os.stat(p).st_size;
    f.append([img,s]);
  return(f);

def partition(arr, low, high):
   i = (low-1)
   pivot = arr[high][1]
   for j in range(low, high):
       if arr[j][1] <= pivot: 
           i = i+1
           arr[i], arr[j] = arr[j], arr[i]
   arr[i+1], arr[high] = arr[high], arr[i+1]
   return (i+1)
 
def quickSort(arr, low, high):
   if len(arr) == 1:
       return arr
   if low < high:
       pi = partition(arr, low, high)
       quickSort(arr, low, pi-1)
       quickSort(arr, pi+1, high)

def linearsearch(arr, x):
  for i in range(len(arr)):
    difference = cv2.subtract(arr[i][0], x);
    b, g, r = cv2.split(difference);
    if cv2.countNonZero(b) == 0 and cv2.countNonZero(g) == 0 and cv2.countNonZero(r) == 0:
      return i;
  return -1

def main():
  t="Enter the number of images you want to insert : ";
  n=int(input(t));
  I=inp_1(n);
  quickSort(I,0,n-1);
  print("All the images have been sorted based on thier size");
  t1="Enter the path of the image you want to search :";
  k=input(t1);
  x=cv2.imread(k);
  N=linearsearch(I,x);
  if N==-1:
    print("Could not find the image");
  else:
    print("The image is found");
    cv2_imshow(I[N][0]);

if __name__ == "__main__":
    main()
```
